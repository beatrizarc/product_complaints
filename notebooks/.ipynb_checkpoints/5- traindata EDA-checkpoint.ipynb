## EDA on preprocessed dataset using TF-IDF (1,2) grams and after selecting top reasons
## on training data

Word cloud: A word cloud can help visualize the most frequent words in the corpus after processing with tf-idf. It is created by generating a cloud of words, where the size of each word is proportional to its frequency in the corpus.

Frequency distribution plot: A plot showing the frequency distribution of words in the corpus can help identify the most frequent words. This plot can be created using a histogram or a bar plot.

Scatter plot: A scatter plot can be used to visualize the distribution of the tf-idf scores for each word in the corpus. This can help identify the words that have high tf-idf scores, indicating their importance in the corpus.

Heatmap: A heatmap can be used to visualize the correlation between words in the corpus after processing with tf-idf. This can help identify groups of words that are highly correlated and may have similar meanings.

Dimensionality reduction techniques: Techniques such as principal component analysis (PCA) and t-SNE can be used to reduce the dimensionality of the tf-idf matrix and visualize it in a lower-dimensional space. This can help identify clusters of documents that are similar to each other based on their tf-idf scores.

```python
# Define a function to plot the most common words for a given label
def plot_most_common_words(label):
    # Filter the data by the given label
    df_label = train_data[train_data['reason']==label]
    # Get the frequency of the words in the complaints for this label
    word_freq = Counter(" ".join(df_label['narrative_tfidf']).split())
    # Compute the top 20 most frequent words
    top_words = pd.Series(word_freq).sort_values(ascending=False)[:20]
    # Plot the top words as a horizontal bar chart
    plt.barh(top_words.index, top_words.values)
    plt.title(label)

# Define the labels to plot
labels= ['Mau Serviço Prestado', 'Condições de entrega', 'Atraso de entrega', 'Enganos']

# Create a figure with subplots for each label
fig, axs = plt.subplots(nrows=1, ncols=len(labels), figsize=(20,5))

# Plot the most common words for each label in a subplot
for i, label in enumerate(labels):
    plt.sca(axs[i])
    plot_most_common_words(label)

# Adjust the layout of the subplots
plt.tight_layout()

# Show the plot
plt.show()
```

## TF-IDF scores

```python
# assuming you have a list of preprocessed texts called `corpus`
# Load the text corpus
corpus = train_data['narrative_tfidf']

vectorizer = TfidfVectorizer(ngram_range=(1,2))
fitted = vectorizer.fit_transform(corpus)

# Get the mapping of terms to feature indices
feature_names = vectorizer.get_feature_names_out()

# Compute the sum of tf-idf scores for each word across all documents
# returns array of scores for each feature len(scores)=len(features)
sum_tfidf_scores = fitted.sum(axis=0).A1

# Sort the indices of the features by their tf-idf scores in descending order
sorted_indices = np.argsort(sum_tfidf_scores)[::-1]

# Get the indices of the top most important features
top_indices = sorted_indices[:20]

# Get the feature names corresponding to the top indices
top_features = [feature_names[i] for i in top_indices]
top_scores=[sum_tfidf_scores[i] for i in top_indices]
for i in range(len(top_features)):
    print(top_features[i],top_scores[i])
```

```python
# for each label
def tfidf_scores(data):
    corpus = data

    vectorizer = TfidfVectorizer(ngram_range=(1,2))
    fitted = vectorizer.fit_transform(corpus)

    # Get the mapping of terms to feature indices
    feature_names = vectorizer.get_feature_names_out()

    # Compute the sum of tf-idf scores for each word across all documents
    # returns array of scores for each feature len(scores)=len(features)
    sum_tfidf_scores = fitted.sum(axis=0).A1

    # Sort the indices of the features by their tf-idf scores in descending order
    sorted_indices = np.argsort(sum_tfidf_scores)[::-1]

    # Get the indices of the top most important features
    top_indices = sorted_indices[:20]

    # Get the feature names corresponding to the top indices
    top_features = [feature_names[i] for i in top_indices]
    top_scores=[sum_tfidf_scores[i] for i in top_indices]
#     for i in range(len(top_features)):
#         print(top_features[i],top_scores[i]) 

    # Return a dictionary of top features and scores
    return {'feature': top_features, 'score': top_scores}
```

```python
# This creates a new DataFrame x containing only the rows where reason is equal to 'Enganos' and the column narrative_tfidf.
enganos=train_data.loc[train_data['reason'] == 'Enganos', 'narrative_tfidf']
atraso=train_data.loc[train_data['reason'] == 'Atraso de entrega', 'narrative_tfidf']
mau_serviço=train_data.loc[train_data['reason'] == 'Mau Serviço Prestado', 'narrative_tfidf']
cond_entrega=train_data.loc[train_data['reason'] == 'Condições de entrega', 'narrative_tfidf']

print('enganos:')
print(tfidf_scores(enganos))
print('\n')
print('atraso entrega:')
print(tfidf_scores(atraso))
print('\n')
print('mau serviço')
print(tfidf_scores(mau_serviço))
print('\n')
print('condiçoes de entrega')
print(tfidf_scores(cond_entrega))
```

```python
# create a dictionary to store top features and scores for each label
top_words = {}
for label in train_data['reason'].unique():
    label_data=train_data.loc[train_data['reason'] == label, 'narrative_tfidf']
    top_words[label] = {'features': tfidf_scores(label_data)['feature'], 'scores': tfidf_scores(label_data)['score']}
    
    
# create a grouped bar plot
fig, ax = plt.subplots(figsize=(8, 8))

bar_width = 0.1
x_pos = np.arange(len(top_words['Enganos']['features']))
colors = ['r', 'b', 'g', 'y', 'm']

for i, label in enumerate(train_data['reason'].unique()):
    ax.bar(x_pos + i*bar_width, top_words[label]['scores'], bar_width, color=colors[i], label=label)

ax.set_xticks(x_pos + bar_width*2)
ax.set_xticklabels(top_words['Enganos']['features'], rotation=90)
ax.set_xlabel('Features')
ax.set_ylabel('TF-IDF Scores')
ax.set_title('TF-IDF Scores for the Top 20 Features in Each Label')
ax.legend()

plt.show()
```

**Importance of a word**: The higher the TF-IDF score of a word, the more important it is in the corpus.  

**Commonality of a word**: The lower the TF-IDF score of a word, the more common it is in the corpus.  

**Relevance of a word**: TF-IDF can be used to find the most relevant words for a particular query or topic. Words with high TF-IDF scores for a particular document or set of documents can be considered the most relevant for that topic. 

**Differences between documents or classes**: TF-IDF can be used to compare the importance of words across different documents or classes. Words with high TF-IDF scores for one document or class, but low TF-IDF scores for others, can be considered distinguishing features for that document or class.

## Clustering

When there's labels, clustering can still be useful:
* Exploration: For example, if you cluster your labeled data and find that documents with a certain label tend to cluster together, it could indicate that these documents have similar characteristics beyond the label. 
* Validation: If you cluster your labeled data and find that documents with the same label are not grouping together, it could indicate that the labels are incorrect or that there is more nuance to the labeling task than originally thought.  
* How? "apply the K-Means algorithm with the number of clusters equal to the number of unique labels and evaluate the clustering using the adjusted rand score, which measures the similarity between the clustering and the true labels. A high adjusted rand score indicates that documents with the same label tend to cluster together"  
* "One way to validate the labels using clustering is to apply clustering to the data with the known labels, and then compare the resulting clusters to the known labels. If the documents with the same label tend to cluster together, then it provides evidence that the labels are accurate. If the documents with the same label are spread out across multiple clusters, then it suggests that the labels may be incorrect or incomplete."
* "Another way to validate the labels using clustering is to apply clustering to the data without the known labels (unsupervised clustering), and then examine the resulting clusters to see if they correspond to meaningful groups. If the resulting clusters contain documents that are similar in some meaningful way (e.g., they discuss similar topics or have similar sentiment), then it provides evidence that the labels are accurate. If the resulting clusters do not correspond to any meaningful groups, then it may suggest that the labels are incorrect or incomplete."
* Labeling new data: For example, if you have a new set of unlabeled documents and you want to assign labels to them, you could first cluster your labeled data and use the clusters as a guide for assigning labels to the new documents.

```python
# Evaluate the clustering by measuring how much documents with the same label are grouped together
ari = adjusted_rand_score(y, clusters)
print("Adjusted Rand Score: ", ari)
```

* The colors of the clusters in the k-means plot and the actual labels plot do not necessarily correspond. In fact, they may be completely different. The colors in each plot represent different things:
 * In the k-means plot, the colors represent the cluster assignments produced by the k-means algorithm.
 * In the actual labels plot, the colors represent the true labels of the data points.
* The purpose of comparing the two plots is to see if the k-means clusters align well with the actual labels, indicating that the clustering algorithm has done a good job of grouping similar documents together according to the true labels.
