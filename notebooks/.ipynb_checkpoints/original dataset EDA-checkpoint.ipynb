```python
import pandas as pd
import numpy as np
from collections import Counter
import nltk
from nltk import word_tokenize
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from sklearn.feature_extraction.text import CountVectorizer
nltk.download('stopwords')
import random
```

```python
df = pd.read_csv("D:\\MS DATA SCIENCE\\product_complaints\\datasets\\data_original.csv")
df.head()
```

```python
lens=[]
for i in range(len(df)):
    tokens=word_tokenize(df['description'].iloc[i])
    lens+=[len(tokens)]
df['description_length'] = lens
```

```python
import seaborn as sns
ax=sns.histplot(x=df['description_length'],binwidth=5)
ax.set(xlabel='Reason',
       ylabel='Complaint length',
       title='Complaint length distribution')
plt.xlim(0, 800)
plt.grid()
plt.tight_layout()
plt.show()
```

```python
print(min(df['description_length']), max(df['description_length']))
```

```python
print('Number of documents:',len(df))

# Get the number of unique words in the complaints
unique_words = set(" ".join(df['description']).split())
print("Number of unique words:", len(unique_words))

# Get the frequency of the words in the complaints
word_freq = Counter(" ".join(df['description']).split())
# print("Frequency of words:", word_freq)

# Get the length of each complaint
complaint_lengths = [len(x.split()) for x in df['description']]
# print("Length of complaints:", complaint_lengths)

# Average length of complaints
print("Avg Length of complaints:", sum(complaint_lengths)/len(df))
```

```python
# Word frequency analysis: This involves identifying the most frequent words in the corpus
# Plot the word frequency distribution
top_words = word_freq.most_common(30)

words = [word[0] for word in top_words]
freqs = [word[1] for word in top_words]

plt.barh(words, freqs)
plt.title("Top 30 most common words")
plt.xlabel("Frequency")
plt.ylabel("Word")
plt.show()
```

```python
# wordcloud ignoring stopwords
stopwords_list = nltk.corpus.stopwords.words('portuguese')
wordcloud = WordCloud(width=800, height=800, stopwords=stopwords_list,background_color='white', max_words=100).generate_from_frequencies(word_freq)
plt.figure(figsize=(8, 8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
# Sample 10,000 complaints randomly because of error
# MemoryError: Unable to allocate 17.0 GiB for an array with shape (34908, 65441) and data type int64
sample_size = 10000
sample_indices = random.sample(range(len(df)), sample_size)
sample_df = df.iloc[sample_indices]

# Create a document-term matrix using CountVectorizer
stopwords_list = nltk.corpus.stopwords.words('portuguese')
vectorizer = CountVectorizer(stop_words=stopwords_list)
doc_term_matrix = vectorizer.fit_transform(sample_df['description'])

# Convert the document-term matrix to a DataFrame
dtm_df = pd.DataFrame(doc_term_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# Compute the top 30 most frequent words
top_words = dtm_df.sum().sort_values(ascending=False)[:30]

# Create a subset of the document-term matrix containing only the top words
dtm_subset = dtm_df[top_words.index]

# Compute the correlation matrix
corr = dtm_subset.corr()

# Create a heatmap of the correlation matrix
sns.heatmap(corr, cmap='coolwarm', annot=False)
```

A heatmap can be used to visualize the correlation between words in the corpus. It shows the similarity between pairs of words based on their co-occurrence in the corpus. The correlation is low.
